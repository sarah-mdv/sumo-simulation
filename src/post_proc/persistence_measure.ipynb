```python
import pandas as pd
import dateutil
import os
import glob
```

```python
path = '../src/output/86400/'
for filename in glob.glob(os.path.join(path, '*.csv')):
    data = pd.read_csv(filename, delimiter=',', header=0, index_col=None,encoding='ascii')
    counts = data["ID"].value_counts()
    counts.to_csv("86400_counts.csv", index=True, header=False, mode='a')
```

```python
data['ID'].nunique()
```

```python
time_s = data.groupby(['ID'])
time_s.tail()
```

```python
c = data['ID'].value_counts()
c
```

```python
c.to_csv("86400_counts.csv", index=True, header=False, mode='a')
```

```python
for n, list in time_s:
    compare = list['Step'].subtract(list['Step'].shift().add(1.0))
    map(lambda x: x != 0, compare)
    d = (list['Step'].groupby(compare.cumsum()).cumcount() + 1).max()
    print(n, d)
```
