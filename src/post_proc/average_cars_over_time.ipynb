```python
import pandas as pd
import dateutil
import os
import glob
import matplotlib as plt
import seaborn as sns
```

```python
def time(x):
    stepmin = x["Step"].min()
    stepmax = x["Step"].max()
    return pd.Series([stepmin, stepmax], index=["First_oc", "Last_oc"])
```

```python
def duration(x):
    start = x.iloc[1][0]
    end = x.iloc[x["Step"].size - 1][0]
    return end - start
```

```python
def to_minutes(x):
    return x/60
```

```python
#Calculate the average number of cars seen per minute over all individual trips
path = '../simulation_output/output'
d = []
duration_data = []
group_durations = []
min_duration = 1
average_cars = pd.DataFrame(columns=["Duration", "Average_cars_per_minute"])
for filename in glob.glob(os.path.join(path, '*.csv')):
    try:
        data = pd.read_csv(filename, delimiter=',', header=0, encoding='ascii',index_col=False )
        total_duration = to_minutes(duration(data))
        if total_duration > min_duration:
            cars_seen = data["ID"].nunique()
            cars_p_minute = cars_seen / total_duration
            print(cars_p_minute)
            average_cars = average_cars.append({"Duration":total_duration, "Average_cars_per_minute":cars_p_minute}, 
                                               ignore_index=True)
    except(OSError):
        print("The file that was expected was not found")
average_cars
```
